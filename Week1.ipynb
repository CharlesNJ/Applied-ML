```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

fruits=pd.read_table('fruit_data_with_colors.txt')
```

```python
fruits.head()
```

```python
fruits
```

```python
#create train-test split
x=fruits[['mass','width','height','color_score']]
y=fruits['fruit_label']

x_train, x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
#plotting 2D - Examining the data
from matplotlib import cm
cmap=cm.get_cmap('gnuplot')
scatter=pd.plotting.scatter_matrix(x_train,c=y_train,marker='o',s=40, hist_kwds={'bins':25},figsize=(15,15),cmap=cmap)
```

```python
#plotting 3D
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x_train['width'],x_train['height'],x_train['color_score'],c=y_train,marker='o',s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```

```python
#identify the labels
lookup_fruit_name = dict(zip(fruits.fruit_label.unique(), fruits.fruit_name.unique()))
lookup_fruit_name
```

```python
#create classifier object
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors =5)
```

```python
#train the classifier
knn.fit(x_train,y_train)
```

```python
#apply the all the instances on the test set
knn.score(x_test,y_test)
```

```python
#predict the new given instances
fruit_prediction=knn.predict([[20,1.6,2.5,5]])
fruit_prediction
lookup_fruit_name[fruit_prediction[0]]
```

```python
#predict the new given instances
fruit_prediction=knn.predict([[1,8000,1005,600]])
fruit_prediction
lookup_fruit_name[fruit_prediction[0]]
```
